```python
pop_pyramid=TNpop.iloc[np.r_[25:44, 50:69]].reset_index(drop=True)
```

```python
pop_pyramid['gender']= 'val'
pop_pyramid.loc[0:19,'gender'] = 'male'
pop_pyramid.loc[19:38,'gender'] = 'female'
pop_pyramid=pop_pyramid.drop(index=[0, 19])
pop_pyramid
```

```python
pop_pyramid.to_csv('../data/pop_pyramid.csv')
```

```python
pop_pyramid['TN count'] = pop_pyramid['TN count'].str.replace(',', '').astype(int)
pop_pyramid['Davidson Co count'] = pop_pyramid['Davidson Co count'].str.replace(',','').astype(int)
```

```python
pop_pyramid.loc[pop_pyramid['gender']=='male', 'TN count'] = pop_pyramid['TN count'].multiply(-1)
```

```python
pop_pyramid.loc[pop_pyramid['gender']=='male', 'Davidson Co count'] = pop_pyramid['Davidson Co count'].multiply(-1)
```

```python
pop_pyramid.to_csv('../data/pop_pyramid.csv')
```
