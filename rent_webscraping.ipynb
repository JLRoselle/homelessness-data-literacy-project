```python
import requests
```

```python
URL = 'https://www.deptofnumbers.com/rent/tennessee/nashville/'

response = requests.get(URL)
```

```python
type(response)
```

```python
response.status_code
```

```python
response.text
```

```python
from bs4 import BeautifulSoup as BS
```

```python
soup=BS(response.text)
```

```python
print(soup.prettify())
```

```python
soup.findAll('table')
```

```python
soup.findAll('table', attrs={'class':'chgtable'})
```

```python
import pandas as pd
from io import StringIO
```

```python
rent=pd.read_html(StringIO(str(soup.find('table', attrs={'class' : 'chgtable'}))))[0]
# Real Gross rent in Nashville in 2019 (USD)
rent
```

```python
vacancy=pd.read_html(StringIO(str(soup.findAll('table', attrs={'class' : 'chgtable'}))))[2]
# Rental Vacancy Rate in Nashville
```

```python
hist_vacancy=pd.read_html(StringIO(str(soup.findAll('table', attrs={'class' : 'stdtable'}))))[2]
# Historical Rental Vacancy Rate for Nashville
```

```python
rent_inc_ratio=pd.read_html(StringIO(str(soup.findAll('table', attrs={'class' : 'chgtable'}))))[3]
# Nashville median annual rent as fraction of median household income
```

```python
hist_rent_inc_ratio=pd.read_html(StringIO(str(soup.findAll('table', attrs={'class' : 'stdtable'}))))[3]
# HIstorical Annual rent as fraction of income in Nashville
```

```python
rent.to_csv('../data/rent.csv')
```

```python
vacancy.to_csv('../data/vacancy.csv')
```

```python
hist_vacancy.to_csv('../data/hist_vacancy.csv')
```
